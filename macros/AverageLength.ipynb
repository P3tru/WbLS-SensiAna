```python
from sympy import *
from sympy.plotting import plot3d
```

```python
x, y, R = symbols('x y R')
```

```python
init_printing()
```

```python
expr3D = 4*pi*(sqrt(x**2+R**2)+sqrt(y**2+R**2))
expr2D = expr3D.subs(y,x)
expr1D = (sqrt(x**2+R**2))
```

```python
subsExpr3D = expr3D.subs(R, 5.7)
subsExpr2D = expr2D.subs(R, 5.7)
subsExpr1D = expr1D.subs(R, 5.7)
```

```python
p1D = plot(subsExpr1D, (x,0,5.7))
p1D.show()
```

```python
intExpr1D = integrate(subsExpr1D, (x,0,5.7))
intExpr1D = intExpr1D/5.7
```
